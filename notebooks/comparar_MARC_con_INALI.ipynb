```python
datos_de_json = []
for dirpath, dirname, filenames in os.walk(INALI_carpeta):
    if not dirname and filenames:
        for filename in filenames:
            filepath = os.path.join(dirpath, filename)
            with open(filepath, 'r') as fpath:
                dato_json = json.loads(fpath.read())
                datos_de_json.append(dato_json)
len(datos_de_json)  # asegurar que el número de variantes sacado del JSON corresponde de lo que sacamos de INALI
```
