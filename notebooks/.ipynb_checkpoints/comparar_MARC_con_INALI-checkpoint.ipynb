```python
import os
import csv
import json
from collections import Counter
```

```python
INALI_carpeta = os.path.join('..', 'agrupaciones_de_INALI')
MARC_ejemplo = os.path.join('..', 'documentos', 'Jonathan_Israel_results_nahuatl.csv')
```

```python
with os.scandir(INALI_carpeta) as inali:
    for entry in inali:
        print(dir(entry))
        print(entry.is_dir())
        print(entry.is_file())
        print(entry.name)
        print(entry.path)
        print()
        if entry.is_file():
            print(entry.name)
            break
```

```python
counter = Counter()
for dirpath, dirname, filenames in os.walk(INALI_carpeta):
    if not dirname and filenames:
        print(dirpath)
        print(dirname)
        print(filenames)
        counter['agrupación'] += 1  # asegurar que la maquina ve a todas las carpetas con JSON file
        print()
print(counter)
```

```python
datos_de_json = []
for dirpath, dirname, filenames in os.walk(INALI_carpeta):
    if not dirname and filenames:
        for filename in filenames:
            filepath = os.path.join(dirpath, filename)
            with open(filepath, 'r') as fpath:
                dato_json = json.loads(fpath.read())
                datos_de_json.append(dato_json)
len(datos_de_json)  # asegurar que el número de variantes sacado del JSON corresponde de lo que sacamos de INALI
datos_de_json[0]
```

```python
# agarrar todos los distintos agrupaciones lingüísticas de los datos de INALI para la búsqueda
agrupaciones = set()
for dato in datos_de_json:
    agrupaciones.add(dato['agrupacion_lingüística'])
print(len(agrupaciones))
# agrupaciones
agrups_lower = [agrupacion.lower() for agrupacion in agrupaciones]
print(len(agrups_lower))
agrups_lower
```

```python
for dato in datos_de_json:
    print(dato['autodenominaciones'])
```

```python
autodenominaciones = [dato['autodenominaciones'] for dato in datos_de_json]
print(len(autodenominaciones))
autodenominaciones[0]
```

```python
variantes = [dato['variante'].lower() for dato in datos_de_json]
print(len(variantes))
variantes
```

```python
# iniciamos nuestro variable para el index del registro MARC como '0'
# ya que en el archivo de hecho los datos empieza con index '1'
registro_index = '0'

# nuestro modelo de datos para los resultados de una búsqueda
# los 'fieldnames' de nuestro archivo de output hay que
# corresponder con los nombres de los llaves en este modelo
datos_del_registro = {
    'index': '',
    'id': '',
    'agrupaciones': [],
    'campos_de_agrupaciones': [],
    'variantes': [],
    'campos_de_variantes': []
}

# inciamos nuestro archivo que guardará resultados positiv0s de una búsqueda
# usamos el modo 'a' en vez de 'w' para que nunca borremos un archivo que ya existe
archivo = os.path.join('..', 'documentos', 'MARC_busqueda_ejemplo.csv')

# chequamos si el archivo ya existe; si existe no queremos escribir el header multiples veces
if os.path.isfile(archivo):
    pass
else:
    with open(archivo, 'a') as outfile:
        # estos fieldnames viene del los nombres de las llaves de nuestro modelo de datos
        fieldnames = ['index', 'id', 'agrupaciones', 'campos_de_agrupaciones',
                      'variantes', 'campos_de_variantes']
        escritor = csv.DictWriter(outfile, fieldnames)
        escritor.writeheader()

# abrimos nuestro archivo de registros MARC
with open(MARC_ejemplo, 'r') as marcs:
    marcs_reader = csv.reader(marcs)
    for row in marcs_reader:
        
        # si el index del row cambia, ya estamos en nuevo registro de MARC
        if row[0] != registro_index and int(row[0]) > 0:
            
            # si el index del registro es 1, no tenemos datos anteriores para escribir al nuestro archivo
            if int(row[0]) > 1:
                # pero si el index es más que 1, chequamos si hay datos encontrados de la búsqueda
                # si no hay resultados registrados de la búsqueda, no hay que escribir nada al archivo de output
                if len(datos_del_registro['agrupaciones']) > 0 or len(datos_del_registro['variantes']) > 0:
                    with open(archivo, 'a') as outfile:
                        fieldnames = ['index', 'id', 'agrupaciones', 'campos_de_agrupaciones',
                                      'variantes', 'campos_de_variantes']
                        escritor = csv.DictWriter(outfile, fieldnames)
                        escritor.writerow(datos_del_registro)

            # reiniciamos nuestros variables para el próximo registro
            datos_del_registro = {
                'index': '',
                'id': '',
                'agrupaciones': [],
                'campos_de_agrupaciones': [],
                'variantes': [],
                'campos_de_variantes': []
            }
            registro_index = row[0]
            datos_del_registro['index'] = registro_index
            continue  # no chequamos los datos de este row porque, como es el row 'LDR', no es relevante
        else:
            if row[1] == '001':  # en el campo 001 tenemos la 'id' del sistema
                datos_del_registro['id'] = row[4]
            else:
                for value in row:
                    if value.lower() in agrups_lower:
                        datos_del_registro['agrupaciones'].append(value)
                        datos_del_registro['campos_de_agrupaciones'].append(row[1])
                    if value.lower() in variantes:
                        datos_del_registro['variantes'].append(value)
                        datos_del_registro['campos_de_variantes'].append(row[1])
```
